```python
import tkinter as tk
from tkinter import *
from tkinter import filedialog,messagebox
from PIL import Image,ImageTk
import os
import cvLib
import configparser
cv2 = cvLib.cv2
np = cvLib.np
```

```python
def mkdir(folder):
    if not os.path.exists(folder):
        os.mkdir(folder)
```

```python
def action_open():
    file_name = filedialog.askopenfilename(initialdir = os.getcwd(),title = "Select file"
                               ,filetypes = (("jpeg files","*.jpg"),("all files","*.*")))
    if file_name == "":
        return
    img = cv2.imread(file_name)
    show_image(lb_frame,img)
    return img
def action_setting():
    create_dlg()
    pass
def action(label):
    if label == "Open":
        action_open()
    elif label == "Setting": 
        action_setting()
        pass
def create_dlg():
    def clicked(act):
        if act == "OK":
            msg = messagebox.askyesnocancel("Save","Do you want to save parameter?")
            if msg is None:
                pass
            elif msg:
                save()
                window.destroy()
                pass
            elif not msg:
                pass 
        elif act == "Cancel":
            window.destroy()
    def save():
        mkdir("parameter")
        config = configparser.ConfigParser()
        config["parameter"]= {"width":str_width.get()
                             ,"height":str_height.get()
                             ,"area":str_area.get()}
        with open('parameter/parameter.cfg', 'w') as configfile:
            config.write(configfile)
            
    window = tk.Toplevel(root)
    window.geometry("300x500+750+50")
    str_width = StringVar()
    str_height = StringVar()
    str_area = StringVar()
    
    tk.Label(window,text="width",width=10).grid(row=0,column=0)
    tk.Label(window,text="height",width=10).grid(row=1,column=0)
    tk.Label(window,text="area",width=10).grid(row=2,column=0)
    
    tk.Entry(window,textvariable=str_width,width=10).grid(row=0,column=1)
    tk.Entry(window,textvariable=str_height,width=10).grid(row=1,column=1)
    tk.Entry(window,textvariable=str_area,width=10).grid(row=2,column=1)
    
    str_width.set("10,100")
    str_height.set("10,100")
    str_area.set("1000,10000")

    tk.Button(window,text="OK",width=10,command=lambda:clicked("OK")).grid(row=3,column=0)
    tk.Button(window,text="Cancel",width=10,command=lambda:clicked("Cancel")).grid(row=3,column=1)
def show_image(frame,image):
    w,h = frame.winfo_width(),frame.winfo_height()
    im_resize = cv2.resize(image,(w,h))
    arr = Image.fromarray(im_resize[...,::-1])
    arr_tk = ImageTk.PhotoImage(arr)
    frame.image = arr_tk
    frame["image"] = arr_tk
    pass

def rect2cvRect(self,rect):
    W,H = lb_frame.winfo_width(),lb_frame.winfo_height()
    H0,W0 = V.image.shape[:2]
    x,y,w,h = rect
    x,y,w,h = int(x*W0/W),int(y*H0/H),int(w*W0/W),int(h*H0/H)
    return [x,y,w,h]
def event(e):
    if e.type == tk.EventType.Motion:
        V.end_point = [e.x,e.y]
        x,y = V.start_point
        w,h = V.end_point[0] - V.start_point[0],V.end_point[1] - V.start_point[1]
        lb_status_frame["text"] = "[x:%d,y:%d,w:%d,h:%d]"%(x,y,w,h)
        lb_status_frame_point["text"] = "[x:%d,y:%d]"%(e.x,e.y)
        if V.bDrawRect = True:
            copy  = self.image.copy()
            if V.end_point[0] < V.start_point[0]:
                self.cvRect = self.rect2cvRect(V.end_point+V.start_point)
            else:
                self.cvRect = self.rect2cvRect(V.start_point+V.end_point)
            x1,y1,x2,y2 = self.cvRect
            cv2.rectangle(copy,(x1,y1),(x2,y2),self.color,self.lw)

            txt = "Rect : [X:{},Y:{},W:{},H:{}]".format(x1,y1,x2-x1,y2-y1)
            self.ui.lb_xywh.setText(txt)

            self.showImage(self.ui.frame,copy)
            pass
        
    elif e.type == tk.EventType.ButtonPress:
        V.start_point = [e.x,e.y]
        V.bDrawRect = True
        pass
    elif e.type == tk.EventType.ButtonRelease:
        V.bDrawRect = False
        pass
    
def bind():
    lb_frame.bind("<Motion>",event)
    lb_frame.bind("<ButtonPress>",event)
    lb_frame.bind("<ButtonRelease>",event)
    pass
class variable:
    def __init__(self):
        self.image = np.ones((640,480,3),np.uint8)
        self.bDrawRect = False
        self.start_point = [0,0]
        self.end_point = [0,0]
```

```python
root = tk.Tk()
root.geometry("640x480+50+50")

f_group_button = tk.Frame(root,bg="turquoise")

lb_frame = tk.Label(root,bg="black")
lb_status_button = tk.Label(root,text="Status",bg="lightgray",borderwidth=3,anchor="nw")
lb_status_frame = tk.Label(root,text="[x,y,w,h]",bg="lightgray",relief="sunken",borderwidth=3,anchor="w")
lb_status_frame_point = tk.Label(root,text="[x,y]",bg="lightgray",relief="sunken",borderwidth=3,anchor="w")
lb_status_progess = tk.Label(root,text="Progess",bg="lightgray",borderwidth=3,anchor="w")

f_group_button.place(relx=0,rely=0,relwidth=0.2,relheight=0.95)
lb_frame.place(relx=0.2,rely=0,relwidth=0.8,relheight=0.95)
lb_status_button.place(relx=0,rely=0.95,relwidth=0.2,relheight=0.05)
lb_status_frame.place(relx=0.2,rely=0.95,relwidth=0.4,relheight=0.05)
lb_status_frame_point.place(relx=0.4,rely=0.95,relwidth=0.2,relheight=0.05)
lb_status_progess.place(relx=0.6,rely=0.95,relwidth=0.4,relheight=0.05)


but_start = tk.Button(f_group_button,text="Start")
but_stop = tk.Button(f_group_button,text="Stop")
but_reset = tk.Button(f_group_button,text="Reset")

but_start.place(relx=0.01,rely=0.95,relwidth=0.325,relheight=0.05)
but_stop.place(relx=0.34,rely=0.95,relwidth=0.325,relheight=0.05)
but_reset.place(relx=0.67,rely=0.95,relwidth=0.325,relheight=0.05)


lst_log = tk.Listbox(f_group_button)
lst_log.place(relx=0,rely=0,relwidth=1,relheight=0.5)


menubar = tk.Menu(root)
menu_file = tk.Menu(menubar, tearoff=0)
menu_file.add_command(label="Open", command=lambda: action("Open"))
menu_file.add_command(label="Save", command=lambda: action("Save"))
menu_file.add_separator()
menu_file.add_command(label="Exit", command=lambda: action("Exit"))
menubar.add_cascade(label="File", menu=menu_file)

menu_setting = tk.Menu(menubar, tearoff=0)
menu_setting.add_command(label="Open Dialog Setting", command=lambda: action("Setting"))
menubar.add_cascade(label="Setting", menu=menu_setting)

menu_font = tk.Menu(menubar, tearoff=0)
menubar.add_cascade(label="Font", menu=menu_font)

menu_lw = tk.Menu(menubar, tearoff=0)
menu_fs = tk.Menu(menubar, tearoff=0)
for i in range(1,11):
    menu_lw.add_command(label="%d"%(i), command=lambda: action("lw%d"%i))
    menu_fs.add_command(label="%d"%(i), command=lambda: action("fs%d"%i))

menu_font.add_cascade(label="Line width", menu=menu_lw)
menu_font.add_cascade(label="Font scale", menu=menu_fs)

bind() # creat binding event
V = variable()
# print(V.image.shape)
show_image(lb_frame,V.image)
#==============================


root.config(menu=menubar)
root.mainloop()
```
